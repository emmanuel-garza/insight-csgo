```python
import pandas as pd
import os
import json
```

```python
# Load the events and save them to a csv file
EVENTS_PATH = '/home/emmanuel/ownCloud/Programming/github-projects/csgo-data/data/events/'

event_dic = {}

for fileobj in os.listdir(EVENTS_PATH):

    file = os.fsdecode(fileobj)

    event_id = file.strip('hltv_event_')
    event_id = event_id.strip('.json')

    with open(EVENTS_PATH + file) as f:
        try:
            data = json.load(f)
            event_dic[event_id] = data
        except:
            continue
```

```python
df_event = pd.DataFrame.from_dict( event_dic, orient='index' )

df_event = df_event.sort_values('dateStart',ascending=False)

df_event[:100]
```

```python
df_event.to_csv('events.csv',index=False)
```
