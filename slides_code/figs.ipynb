```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

 
sns.set_style("whitegrid")
sns.set(font_scale=6)


mpl.rcParams.update({'text.color' : "gray",
                     'axes.labelcolor' : "gray"})

N = 1000

# Dataset
df=pd.DataFrame({'X': np.random.randn(N)*N+range(1,N+1), 'Y': np.random.randn(N)*N+range(1,N+1), 'Z': (np.random.randn(N)*N+range(1,N+1))*2 })
 
# plot
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X']/df['X'].max(), df['Y']/df['Y'].max(), df['Z']/df['Z'].max(), c=df['Z']/df['Z'].max(), s=320, linewidths=3, edgecolors='k')

ax.set_xlabel('Rating',labelpad=40)
ax.set_ylabel('Prize Rating',labelpad=40)
ax.set_zlabel('HS %',labelpad=40)

# sns.despine(left=True)
# sns.despine(right=True)
# sns.despine(top=True)
# sns.despine(fig=fig, ax=ax, top=False, right=False, left=False, bottom=False, offset=None, trim=False)
# ax.set_frame_on(False)
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_zticklabels([])

# ax.set_facecolor('gray')

ax.view_init(30, 120)




#plt.savefig('3d.png', transparent=True)

plt.show()


# plot
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(111)
ax.scatter(df['X']/df['X'].max(), df['Y']/df['Y'].max(), 
    c=df['X']/df['X'].max()*(df['Y']/df['Y'].max()), 
    s=320, linewidths=3, edgecolors='k')

ax.set_yticklabels([])
ax.set_xticklabels([])

ax.set_xlabel('Rating',labelpad=40)
ax.set_ylabel('HS %',labelpad=40)
ax.set_facecolor('gray')


#plt.savefig('2d.png', transparent=True)

plt.show()

```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.style as style
import numpy as np
import pandas as pd
import seaborn as sns

#data = {'Place':['1st','2nd','3rd','4th'], 'Esport':[400,400,320,240], 'NHL':[400,325,310,300]}
data = {'Place':['1st','2nd','3rd','4th','1st','2nd','3rd','4th'], 'Type':['Esport','Esport','Esport','Esport','NHL','NHL','NHL','NHL'],
'Valuation':[400,400,320,240,400,325,310,300]}
df = pd.DataFrame(data)

df.head()

plt.style.use('fivethirtyeight')

# sns.set_style("whitegrid")
sns.set(font_scale=6)


mpl.rcParams.update({'text.color' : "gray",
                     'axes.labelcolor' : "white",
                     'xtick.color' : "white"})

# plot
fig = plt.figure(figsize=(20,16))

ax = sns.barplot(y='Place',x='Valuation',hue='Type',data=df,order=['1st','2nd','3rd','4th'],palette=['#9000EE','#5EEE00'],alpha=0.9)

ax.set_yticklabels([])
ax.set_xticklabels(['0','100','200','300','400',])


ax.set_ylabel('')
ax.set_facecolor('gray')

ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
ax.set_xlabel('Valuation in $ millions')

plt.savefig('valuation.svg', transparent=True)

 
```
